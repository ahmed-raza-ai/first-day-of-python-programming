```python

class Customer:
  def __init__(self, name , id ):
    self.name = name
    self.id = id

  def displayinfo(self):
    print("the name of the persom is " , self.name)
    print("the id of the person is " , self.id )

class Credit_card_payment  (Customer):

  def __init__( self,name ,id ,balance ,pin):
    super().__init__( name , id )
    self.balance = balance
    self.__pin = pin

  def __verifypin( self ,inputpin ):
     return self.__pin == inputpin

  def deposit( self ,amount):
    if amount > 0:
     self.balance += amount
     print("your deposit is sucessfull")
     print("your new balance is" ,self.balance)
    else:
      print("invalid amount")

  def withdraw(self ,amount ,inputpin):
    if self.__verifypin(inputpin):
     if amount > 0 and amount <= self.balance:
      self.balance -= amount
      print("your withdraw is sucesfull")
      print("your new balance is " , self.balance)
     else:
      print("your current balance is low ")
    else:
      print(" wrong pin acess denied")

  def display_balance(self):
    print("your balance is " , self.balance)

  def showinfo(self):
   self.displayinfo()
   print("your current balance is " ,self.balance)

c1 = Credit_card_payment("MUHAMMAD AHMED RAZA" , 67028 , 5000 , 1234)

c1.deposit(1000)
c1.display_balance()
c1.withdraw(5000 , 1234)
c1.showinfo()
c1.deposit(2000)
c1.withdraw(10000 ,1234)
c1.display_balance()





```
