```python
class Customerinfo:
  def __init__(self,name ,id ,location):
    self.__name = name
    self.__id = id
    self.__location = location

  def get__name(self):
      return self.__name
  def get__id(self):
      return self.__id
  def get__location(self):
     return self.__location


c1 = Customerinfo("ahmed", 67028 ,"karachi")
print("the name of the customer is" , c1.get__name())
print("the id of the customer is" , c1.get__id())
print("the location of the customer is" , c1.get__location())
```

```python
class info:
  def __init__(self ,age):
    self.__age = age
    if self.__age > 0:
      self.__age = age
      print("the age of the person is valid")
    else:
      print("invalid age please enter age again")
  def get__age(self):
    return self.__age
person = info(-1)


print("the age of the person is " , person.get__age())

```
